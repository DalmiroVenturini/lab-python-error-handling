# Step 1: Define the function for initializing the inventory with error handling
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory

````python
#2
def calculate_total_price(products):
    total_price = 0.0
    for product in products:
        valid_price = False
        while not valid_price:
            try:
                price = float(input(f"Enter the price for {product}: "))
                if price < 0:
                    raise ValueError("Price cannot be negative! Please enter a valid price.")
                valid_price = True
                total_price += price
            except ValueError as error:
                print(f"Error: {error}")
    return total_price
````

````python
#3
def get_customer_orders(inventory):
    orders = {}
    valid_order_count = False
    
    #Primero manejamos el número de órdenes
    while not valid_order_count:
        try:
            num_orders = int(input("Enter the number of orders: "))
            if num_orders < 0:
                raise ValueError("The number of orders cannot be negative. Please enter a valid number.")
            valid_order_count = True
        except ValueError as error:
            print(f"Error: {error}")

    #Luego procesamos cada orden
    for _ in range(num_orders):
        valid_product = False
        while not valid_product:
            product_name = input("Enter the product name: ").strip()   #eliminando espacios en blanco al inicio y al fin
            if product_name in inventory and inventory[product_name] > 0:
                valid_quantity = False
                while not valid_quantity:
                    try:
                        quantity = int(input(f"Enter the quantity for {product_name}: "))
                        if quantity < 0:
                            raise ValueError("Quantity cannot be negative.")
                        if quantity > inventory[product_name]:
                            raise ValueError("Not enough stock available.")
                        orders[product_name] = quantity
                        valid_quantity = True
                        valid_product = True
                    except ValueError as error:
                        print(f"Error: {error}")
            else:
                print(f"Error: '{product_name}' is not available or out of stock.")
    
    return orders

````
